```python
# it will return column name along with total values which are missing 
df.isnull().sum()
```

```python
# now will remove columns which are null from table for eg. country, agent, company they are huge in numbers
# df.drop is used to remove column or row use axis = 0 for row and axis = 1 for col
# to make changes in database pass inplace = true
# use dropna to remove only null values from table
```

```python
df.drop(['company', 'agent'], axis = 1, inplace = True)
df.dropna(inplace = True)
```

```python
df = df[df['adr']< 5000]
```

```python
# Data Analysis and Visualizations
# we need to analyze how much is the percentage of cancellation of bookings to see if issue is minor or major
```

```python
canceled_percent = df['is_canceled'].value_counts(normalize = True)
print(canceled_percent)

# doing visualization of data
plt.figure(figsize = (5, 4))
plt.title('Reservation status count')
plt.bar(['Not canceled', 'Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.7)
plt.show()
```

```python
# Depending on the hotel (city hotels and resort hotels) we will find out whose cancellation rate is bigger
# in numbers will use countplot for visualization
# 
```

```python
plt.figure(figsize = (8, 4))
axl = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = "Blues")
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor = (1, 1))
plt.title("Reservation status in different hotels", size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservations') 
plt.show()
```

```python
# now we want our visuals in percentage of cancellations
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# we want to know is price can be the reason of more cancellations in city hotels as cancellation rate of city hotels is almost 50%
```

```python
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
plt.figure(figsize = (15, 6))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 15)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 8)
plt.show()
```

```python
# we want to see which month has highest reservations in hotel booking
```

```python
df['month'] = df['reservation_status_date'].dt.month
plt.figure(figsize = (13, 6))
axl = sns.countplot(x = 'month', hue = 'is_canceled',  data = df, palette = 'bright')
legend_labels,_ = axl.get_legend_handles_labels()
axl.legend(bbox_to_anchor = (1, 1))
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')
plt.legend(['not canceled', 'canceled'])
plt.show()
```

```python
# ploting the average daily rate for each month
# here we are only seeing data for canceled reservations because we want to analyze 
# that price is the factor that contributed toward the booking cancellations
```

```python
plt.figure(figsize = (12, 6))
plt.title('ADR per month', fontsize = 24)
# Use 'x' and 'y' keyword arguments for the columns
sns.barplot(x = 'month', y='adr', data = df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index())
plt.show()           
```

```python
# we will analyze which country has higher cancellations 
```

```python
canceled_data = df[df['is_canceled'] == 1]
top_10_country = canceled_data['country'].value_counts()[:10]
plt.figure(figsize = (7, 7))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

```python
# now will check the daily average rate of cancellations
```

```python
canceled_df_adr = canceled_data.groupby('reservation_status_date')[['adr']].mean()
canceled_df_adr.reset_index(inplace = True)
canceled_df_adr.sort_values('reservation_status_date', inplace = True)

not_canceled_data = df[df['is_canceled'] == 0]
not_canceled_df_adr = not_canceled_data.groupby('reservation_status_date')[['adr']].mean()
not_canceled_df_adr.reset_index(inplace = True)
not_canceled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20, 6))
plt.title('Average Daily Rate')
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label = 'not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label = 'canceled')
plt.legend()
plt.show()
```

```python
# there is lot of inconsistancy in data before 2016 data is missing so will fliter out
# data and consider only available data for analysis
```

```python
canceled_df_adr = canceled_df_adr[(canceled_df_adr['reservation_status_date'] > '2016') & (canceled_df_adr['reservation_status_date'] < '2017-09')]
not_canceled_df_adr = not_canceled_df_adr[(not_canceled_df_adr['reservation_status_date'] > '2016') & (not_canceled_df_adr['reservation_status_date'] < '2017-09')]
```

```python
plt.figure(figsize = (15, 6))
plt.title('Average Daily Rate', fontsize = 20)
plt.plot(not_canceled_df_adr['reservation_status_date'], not_canceled_df_adr['adr'], label = 'not canceled')
plt.plot(canceled_df_adr['reservation_status_date'], canceled_df_adr['adr'], label = 'canceled')
plt.legend(fontsize = 20)
plt.show()
```
